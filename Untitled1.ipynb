```python
def parse_yes_no(value):
    print(f"\nProcessing value: {repr(value)}")  # Debug print
    
    # Handle missing values
    if pd.isna(value):
        print("-> Missing value, returning No")
        return 'No'
    
    # Check if already Yes/No
    if isinstance(value, str):
        value_lower = value.lower().strip()
        if value_lower in ['yes', 'no']:
            print(f"-> Already Yes/No: {value}")
            return value.capitalize()
    
    # HTML parsing logic
    if isinstance(value, str):
        print("-> Checking HTML patterns...")
        
        # Check for Yes pattern
        yes_match = re.search(r'<span\s+class\s*=\s*["\']catalogue_circle\s+-yes["\']', value)
        if yes_match:
            print(f"-> Found YES pattern at positions {yes_match.span()}")
            return 'Yes'
        
        # Check for No pattern
        no_match = re.search(r'<td\s+class\s*=\s*["\']custom_table-heading_general["\']', value)
        if no_match:
            print(f"-> Found NO pattern at positions {no_match.span()}")
            return 'No'
    
    print("-> No patterns matched, returning Unknown")
    return 'Unknown'
```

```python
from bs4 import BeautifulSoup
import pandas as pd

def parse_remote_support(html):
    """
    Parses HTML table rows to determine Remote Support status
    Returns "Yes" if <span class="catalogue_circle -yes"> exists, else "No"
    """
    if not isinstance(html, str) or not html.strip():
        return "No"
    
    soup = BeautifulSoup(html, 'html.parser')
    tr_tags = soup.find_all('tr', {'data-entity-id': True})
    
    results = []
    for tr in tr_tags:
        # Find the specific td containing remote support indicator
        support_td = tr.find('td', class_='custom_table-heading_general')
        if support_td:
            yes_span = support_td.find('span', class_='catalogue_circle -yes')
            results.append("Yes" if yes_span else "No")
        else:
            results.append("No")
    
    return results if results else ["No"]

# Example usage:
html = '''your_full_html_content_here'''
support_statuses = parse_remote_support(html)
print(support_statuses)
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Load your CSV file
df = pd.read_csv('assessment_details.csv')

def check_support(html_content):
    """
    Correctly navigates:
    div.custom__table-responsive â†’ table â†’ tbody â†’ tr â†’ td.custom__table-heading__general â†’ span.catalogue__circle.-yes
    """
    if pd.isna(html_content) or not isinstance(html_content, str):
        return "No"
    
    try:
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # Navigate through the full HTML structure
        table_div = soup.find('div', class_='custom__table-responsive')
        if not table_div:
            return "No"
            
        table = table_div.find('table')
        if not table:
            return "No"
            
        tbody = table.find('tbody')
        if not tbody:
            return "No"
            
        # Process each row
        for row in tbody.find_all('tr'):
            # Find the specific td within this row
            target_td = row.find('td', class_='custom__table-heading__general')
            
            if target_td:
                # Check for yes span (note the class format)
                if target_td.find('span', class_='catalogue__circle -yes'):
                    return "Yes"
                    
    except Exception as e:
        print(f"Error parsing HTML: {e}")
    
    return "No"

# Apply to both support columns
df['Remote_Support'] = df['Remote_Support'].apply(check_support)
df['Adaptive_Support'] = df['Adaptive_Support'].apply(check_support)

# Verify results
print("First 5 rows after processing:")
print(df[['Name', 'Remote_Support', 'Adaptive_Support']].head())

# Save results
df.to_csv('processed_assessment_details.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Setup browser
driver = webdriver.Chrome()
driver.get("https://www.shl.com/solutions/products/product-catalog/")

all_data = []
headers = ["Individual Test Solutions", "Remote Testing", "Adaptive/IRT", "Test Type", "Description", "Duration", "Job-Level", "Languages", "URL"]
first_page_completed = False  # Track first page completion

while True:
    time.sleep(2)
    
    # Get all tables
    tables = driver.find_elements(By.CSS_SELECTOR, ".custom__table-responsive")
    
    # Select which table to scrape
    if not first_page_completed:
        # First page - use second table if available
        if len(tables) >= 2:
            table = tables[1]
            target_pagination = 1  # Use second pagination
        else:
            table = tables[0]
            target_pagination = 0
        first_page_completed = True
    else:
        # Subsequent pages - always use first table
        table = tables[0]
        target_pagination = 0
    
    # Get pagination container
    paginations = driver.find_elements(By.CSS_SELECTOR, ".pagination")
    if len(paginations) <= target_pagination:
        print("Pagination container not found")
        break
    
    # Get rows (skip header if we already have headers)
    rows = table.find_elements(By.TAG_NAME, "tr")
    if len(all_data) == 0:  # Only skip header on first run
        rows = rows[1:]  # Skip header
    
    # Collect data
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, "td")
        if cols:
            row_data = []
            product_url = ""
            
            # Process each column in the row
            for col in cols:
                # Check if this is a "yes" indicator cell
                yes_indicator = col.find_elements(By.CSS_SELECTOR, "span.catalogue__circle.-yes")
                if "custom__table-heading__general" in col.get_attribute("class") and yes_indicator:
                    row_data.append("yes")
                else:
                    row_data.append(col.text.strip())
                
                # Try to find product URL in the first column
                if col == cols[0]:
                    # Look for any links in this column
                    links = col.find_elements(By.TAG_NAME, "a")
                    if links:
                        product_url = links[0].get_attribute("href")
            
            # Add placeholder values for the additional columns
            row_data.extend(["", "", "", "", product_url])  # Description, Duration, Job-Level, Languages, URL
            
            all_data.append(row_data)
    
    print(f"Collected {len(rows)} rows from {'second' if not first_page_completed else 'first'} table")
    
    # Click the correct next button
    try:
        next_btn = paginations[target_pagination].find_element(
            By.CSS_SELECTOR, 
            'li.pagination__item.-arrow.-next a.pagination__arrow'
        )
        next_btn.click()
    except:
        print("No more pages or pagination failed")
        break

# Save and display data
if all_data:
    df = pd.DataFrame(all_data, columns=headers)
    df.to_csv('correct_individual_tests.csv', index=False)
    print(f"\nSuccess! Saved {len(df)} rows")
    
    # Display all data in Jupyter
    pd.set_option('display.max_rows', None)
    display(df)
else:
    print("No data collected")

driver.quit()
```

```python
# Import necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random

# Load the CSV file
df = pd.read_csv('correct_individual_tests.csv')

# Function to extract job description from a URL
def extract_job_description(url):
    try:
        # Add a random delay to avoid being blocked
        time.sleep(random.uniform(1, 3))
        
        # Send a GET request to the URL
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Find the job description
            # Note: You'll need to adjust the selector based on the actual structure of the job pages
            # This is a common pattern, but you might need to inspect the pages and update this
            job_description = soup.find('div', class_='job-description')
            
            if job_description:
                return job_description.get_text(strip=True)
            else:
                # Try alternative selectors
                job_description = soup.find('div', id='job-description')
                if job_description:
                    return job_description.get_text(strip=True)
                else:
                    # Another common pattern
                    job_description = soup.find('div', class_='description')
                    if job_description:
                        return job_description.get_text(strip=True)
                    else:
                        return "Description not found. Check the page structure."
        else:
            return f"Failed to retrieve page: Status code {response.status_code}"
    except Exception as e:
        return f"Error: {str(e)}"

# Create a new column or update existing job description column
df['job_description'] = None

# Process each URL and update the dataframe
for index, row in df.iterrows():
    # Get the URL from the appropriate column
    # Update 'url_column_name' to match your actual column name
    url = row['URL']  # Replace with your actual URL column name
    
    # Print progress
    print(f"Processing {index+1}/{len(df)}: {url}")
    
    # Extract job description
    job_description = extract_job_description(url)
    
    # Update the dataframe
    df.at[index, 'job_description'] = job_description
    
    # Save after every 10 records (to avoid losing progress if script crashes)
    if (index + 1) % 10 == 0:
        df.to_csv('correct_individual_tests_with_descriptions.csv', index=False)
        print(f"Progress saved at {index+1} records")

# Save the final dataframe
df.to_csv('correct_individual_tests_with_descriptions.csv', index=False)

print("Job description extraction complete!")
```
